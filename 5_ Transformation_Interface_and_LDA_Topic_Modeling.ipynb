```python
from gensim import corpora, models, similarities
from gensim.models.ldamodel import LdaModel
import os
from pprint import pprint
if (os.path.exists("/tmp/dictionary_today.dict")):
    dictionary = corpora.Dictionary.load('/tmp/dictionary_today.dict')
    corpus = corpora.MmCorpus('/tmp/corpus_today.mm')
    print("Load Dictionary and Corpus : Success")
else:
    print("Please following instruction : 4_Build_Dictionary_and_Corpus")
```

## Creating a transformation

```python
tfidf = models.TfidfModel(corpus)
```

```python
print(tfidf)
```

## Transformation Vector

```python
new_doc = "raffi ahmad ahok jokowi"
new_vec = dictionary.doc2bow(new_doc.lower().split())
# pprint(new_vec)

entity = new_doc.lower().split()
entity = enumerate(entity)
vector_idf = tfidf[new_vec]
# vector_idf = enumerate(vector_idf)
for id, person in entity:
    print(person)
    print('tf : %s' % (new_vec[id][1]))
    print('tf-idf : %s' % (vector_idf[id][1]))
    print('=================')
```

```python
num_topics = 20
chunksize = 20
passes = 20
iterations = 200
eval_every = None


temp = dictionary[0]  
id2word = dictionary.id2token

%time model = LdaModel(corpus=corpus, id2word=id2word, chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, num_topics=num_topics, \
                       passes=passes, eval_every=eval_every)
```

```python
%time top_topics = model.top_topics(corpus, num_words=20)

avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)
```
